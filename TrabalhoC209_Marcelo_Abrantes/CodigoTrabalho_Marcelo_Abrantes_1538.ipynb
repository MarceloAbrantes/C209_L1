**Definição:**

A subtração de fundo (background subtraction) é uma técnica comum em visão computacional que visa isolar objetos de interesse em uma imagem, removendo o fundo que os envolve. Essa técnica é amplamente utilizada em diversas aplicações, incluindo rastreamento de objetos, detecção de movimento, segurança, realidade aumentada, entre outras.

 O código realiza a subtração de fundo condicional entre uma imagem original e um fundo fornecido pelo usuário.

No código foi utilizado a biblioteca OpenCv que foi instalada com o código:

**pip install opencv-python**.

Passo a passo do código:


* **1.** Importação de Bibliotecas:
  
cv2 é a biblioteca OpenCV usada para manipulação de imagens.

numpy é usada para operações numéricas em matrizes
  .
* **2.** Carregando Imagens

As imagens originais e de fundo são carregadas usando a função cv2.imread.

* **3.** Redimensionando o Fundo

O fundo é redimensionado para que tenha o mesmo tamanho da imagem original. Isso é feito usando a função cv2.resize.

* **4.** Subtração de Fundo

A subtração de fundo é feita calculando a diferença pixel a pixel entre a imagem original e o fundo. Isso é feito com a função cv2.absdiff.  
 
* **5.** Conversão para Escala de Cinza

A imagem resultante da subtração de fundo é convertida para escala de cinza usando cv2.cvtColor.
  
* **6.** Criação da Máscara Binária

máscara binária destaca os objetos em branco usando a função cv2.threshold.

* **7.** Aplicação da Máscara

isola os objetos da imagem original, tornando o fundo preto usando a função cv2.bitwise_and.

* **8.** Exibição da Imagem Final

A imagem final, que contém apenas os objetos isolados do fundo, é exibida usando cv2.imshow.
  
* **9.** Espera do Usuário

uma tecla para fechar a janela da imagem final.
  
* **10.** Fechando Janelas

as janelas abertas são fechadas usando cv2.destroyAllWindows.

## Código

```python
import cv2
import numpy as np

# Carregue a imagem original e o fundo fornecido pelo usuário
imagem_original = cv2.imread('imagem_original.jpg')
fundo = cv2.imread('fundo.jpg')

# Redimensione o fundo para que tenha o mesmo tamanho da imagem original
fundo = cv2.resize(fundo, (imagem_original.shape[1], imagem_original.shape[0]))

# Subtraia o fundo da imagem original
imagem_sem_fundo = cv2.absdiff(imagem_original, fundo)

# Converta a imagem resultante para escala de cinza
imagem_sem_fundo_cinza = cv2.cvtColor(imagem_sem_fundo, cv2.COLOR_BGR2GRAY)

# Aplique uma máscara binária para isolar os objetos do fundo
_, mascara = cv2.threshold(imagem_sem_fundo_cinza, 30, 255, cv2.THRESH_BINARY)

# Aplique a máscara na imagem original para obter a imagem final sem fundo
imagem_final = cv2.bitwise_and(imagem_original, imagem_original, mask=mascara)

# Exiba a imagem final
cv2.imshow('Imagem Sem Fundo', imagem_final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
